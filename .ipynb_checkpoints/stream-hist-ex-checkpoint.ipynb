# Histograms in Real-time

## with Plotly's Streaming API

First import a few module

```python
import plotly.plotly as py
import plotly.tools as tls
```

```python
from plotly.graph_objs import Figure, Data, Layout
from plotly.graph_objs import Stream
from plotly.graph_objs import Histogram
```

```python
import numpy as np
import time
```

Next, retrieve a stream id from your credentials file and make instance of Stream id object:

```python
# Get stream ids in credentials file
my_stream_ids = tls.get_credentials_file()['stream_ids']

# Use for stream id
my_stream_id = my_stream_ids[0]

# Make instance of stream id object 
my_stream = Stream(token=my_stream_id) # N.B. link stream id to 'token' key
```

Every Plotly trace object has a `'stream'` key available to initialize a data stream.

The list of available keys in each graph object can be viewed by running help. So, for example:

```python
help(Histogram)  # help()

# See that Histogram has a 'stream' key
```

Next, let's initialize a simple figure object and send it to Plotly via `py.iplot()`.

Please note that (this is possibly the reason why your try was unsuccesful):

> The `Histogram` graph object must be initialized with a non-empty `'x'` (or `'y'` key) for the stream to work, unlike for the `Scatter` graph object.

Then, 

```python
# Initialize trace of streaming plot by embedding the unique stream_id
my_data = Data([Histogram(x=[1],    # init. values (N.B. must be non-empty here!!!)
                          stream=my_stream)])      # N.B. embed stream id, 1 per trace
```

```python
my_layout = Layout(title='Streaming histograms!', # plot title
                   showlegend=False,              # remove legend 
                  )
```

```python
# Make instance of figure object
my_fig = Figure(data=my_data, layout=my_layout)

# Initialize streaming plot, show result in notebook
py.iplot(my_fig, filename='stream-hist-ex1')
```

Next, open the stream via the stream link object:

```python
# Make instance of the Stream link object, 
#  with same stream id as Stream id object
s = py.Stream(my_stream_id)

# Open the stream
s.open()
```

And stream some data:

```python
N = 20
i = 0

while i<N:
    
    s_data = dict(x=np.random.randn(100))  # 100 normaly distributed values

    s.write(s_data) # send data to stream once per second,
    time.sleep(1)   #   see results in cell above
    
    i += 1
    
# Close stream when done
s.close()
```
