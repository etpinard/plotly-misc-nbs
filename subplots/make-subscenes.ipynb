# Plotly 3d subplot generation

<br>
<br>

```python
import plotly
plotly.__version__
```

```python
import plotly.plotly as py   # sign in with your credentials file
import plotly.tools as tls
from plotly.graph_objs import *
```

```python
import numpy as np

# Some random number generating function
def get_random_data(N=50): 
    return np.random.random(N), np.random.random(N), np.random.random(N)
```

```python
# Make figure object with a 2x2 sub-scene layout
fig = tls.make_subplots(rows=2, cols=2,
                        specs=[[{'is_3d': True}, {'is_3d': True}],
                               [{'is_3d': True}, {'is_3d': True}]])
```

```python
# Append trace by row & col to fig object
for row in range(2):
    for col in range(2):
        x, y, z = get_random_data()
        fig.append_trace(Scatter3d(
            x=x,
            y=y,
            z=z,
            mode='markers'
        ), row+1, col+1)

# Add title
layout = Layout(
    title='3D suplot generation!',
)
```

```python
# The figure object is:
print(fig.to_string())
```

```python
# Send figure object to plotly!
py.iplot(fig, filename='3d-subplot-generation')
```
