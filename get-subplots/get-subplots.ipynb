# tools.get_subplots

## new and improved

```python
import plotly.plotly as py
import plotly.tools as tls
from plotly.graph_objs import *
```

```python
# Some generic trace
trace1 = Scatter(
    x=[1,2,3,4,5],
    y=[2,1,2,1,2]
)

from copy import deepcopy

# Helper function binding trace to a x,y axis pair
def bind(trace, x_num, y_num):
    trace['xaxis'] = 'x{0}'.format(x_num)
    trace['yaxis'] = 'y{0}'.format(y_num)
    return deepcopy(trace)
```

### Backward compatibility break 

Note that the horizontal and vertical spacing between subplots are now computed using the number of columns and rows respectively (instead of being fixed for all `rows` and `columns` values)

The old way of doing things yielded spurious results (e.g. domains < 0 and/or > 1) when the number of rows [columns] was greater than 1 / (horizontal_spacing) [1 / (vertical_spacing)].

More specifically, the `tls.get_subplots` keyword arguments are now:

```python
help(tls.get_subplots)
```

Note that `horizontal_spacing` and `vertical_spacing` can be set as before, only their default has been modified.

### Shared axis options

```python
fig = tls.get_subplots(rows=4, columns=1, 
                       shared_xaxes=True,  # NEW !!!
                       print_grid=True)
```

```python
fig['data'] += [bind(trace1, 1, 1)]
fig['data'] += [bind(trace1, 1, 2)]
fig['data'] += [bind(trace1, 1, 3)]
fig['data'] += [bind(trace1, 1, 4)]
```

```python
print fig.to_string()  # notice 'anchor': 'free'
```

```python
py.iplot(fig, filename='get_subplots-shared')
```

```python
fig = tls.get_subplots(rows=1, columns=4, 
                       shared_yaxes=True,  # sharing y-axes also works!
                       print_grid=True)
```

### Custom layout using subplot specs

Inspired by matplotlib's [gridspec](http://matplotlib.org/users/gridspec.html).

Here are a few examples:

```python
fig = tls.get_subplots(rows=3, columns=3,
                       specs=[[{}, {}, {'rowspan': 2}],
                              [{'colspan': 2}],
                              [{'colspan': 3}]],  # specs w/ 'colspan' & 'rowspan'
                       print_grid=True)
```

```python
fig['data'] += [bind(trace1, 1, 1)]
fig['data'] += [bind(trace1, 2, 2)]
fig['data'] += [bind(trace1, 3, 3)]
fig['data'] += [bind(trace1, 4, 4)]
fig['data'] += [bind(trace1, 5, 5)]
```

```python
py.iplot(fig, filename='get_subplots-specs')
```

```python
fig = tls.get_subplots(rows=2, columns=3,
                       specs=[[{}, {}, {}],
                              [{}]],
                       print_grid=True)
```

```python
fig['data'] += [bind(trace1, 1, 1)]
fig['data'] += [bind(trace1, 2, 2)]
fig['data'] += [bind(trace1, 3, 3)]
fig['data'] += [bind(trace1, 4, 4)]
```

```python
py.iplot(fig, filename='get_subplots-specs2')
```

```python
fig = tls.get_subplots(rows=3, columns=3,
                       specs=[[{'rowspan': 3}, {}, {}],
                              [{'isEmpty': True}, {}, {}],
                              [{'isEmpty': True}, {'colspan': 2}]],
                       print_grid=True)
```

I'm not too existed about `'isEmpty'` but I think it is more intuitive than having `'rowspan'` magically fill the columns above. 

Thoughts?

```python
fig['data'] += [bind(trace1, 1, 1)]
fig['data'] += [bind(trace1, 2, 2)]
fig['data'] += [bind(trace1, 3, 3)]
fig['data'] += [bind(trace1, 4, 4)]
fig['data'] += [bind(trace1, 5, 5)]
fig['data'] += [bind(trace1, 6, 6)]
```

```python
py.iplot(fig, filename='get_subplots-specs4')
```

```python
fig = tls.get_subplots(rows=2, columns=2,
                       horizontal_spacing=0, vertical_spacing=0,
                       specs=[[{'l': 0.1}, {'b': 0.2}],
                              [{'t': 0.2}, {'r': 0.1}]],  # custom padding !
                       print_grid=True)
```

```python
fig['data'] += [bind(trace1, 1, 1)]
fig['data'] += [bind(trace1, 2, 2)]
fig['data'] += [bind(trace1, 3, 3)]
fig['data'] += [bind(trace1, 4, 4)]
```

```python
py.iplot(fig, filename='get_subplots-specs5')
```
