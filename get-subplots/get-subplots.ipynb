# tools.make_subplots

## a new and improved get_subplots

```python
import plotly.plotly as py
import plotly.tools as tls
from plotly.graph_objs import *
```

```python
# Some generic trace
trace1 = Scatter(
    x=[1,2,3,4,5],
    y=[2,1,2,1,2]
)

from copy import deepcopy

# Helper function binding trace to a x,y axis pair
def bind(trace, x_num, y_num):
    trace['xaxis'] = 'x{0}'.format(x_num)
    trace['yaxis'] = 'y{0}'.format(y_num)
    return deepcopy(trace)
```

### Bug fix in get_subplots

Note that the horizontal and vertical spacing between subplots are now computed using the number of columns and rows respectively (instead of being fixed for all `rows` and `columns` values)

The old way of doing things yielded spurious results (e.g. domains < 0 and/or > 1) when the number of rows [columns] was greater than 1 / (horizontal_spacing) [1 / (vertical_spacing)].


Note that `horizontal_spacing` and `vertical_spacing` can be set as before, only their default has been modified.

### Shared axis options

##### Important question

If `shared_xaxes=True` (and `start_cell='top-left'` our new default), should the shared xaxes be on the top or bottom row ?

```python
fig = tls.make_subplots(rows=4, cols=1, 
                        shared_xaxes=True,  # NEW !!!
                        print_grid=True)
```

```python
fig['data'] += [bind(trace1, 1, 1)]
fig['data'] += [bind(trace1, 1, 2)]
fig['data'] += [bind(trace1, 1, 3)]
fig['data'] += [bind(trace1, 1, 4)]
```

```python
print fig.to_string()  # notice 'anchor': 'free'
```

```python
py.iplot(fig, filename='get_subplots-shared')
```

Or the shared axes always be at the bottom like it is if `start_cell='bottom-left'`:

```python
fig = tls.make_subplots(rows=4, cols=1, 
                        shared_xaxes=True,  # NEW !!!
                        start_cell='bottom-left',
                        print_grid=True)
```

```python
fig['data'] += [bind(trace1, 1, 1)]
fig['data'] += [bind(trace1, 1, 2)]
fig['data'] += [bind(trace1, 1, 3)]
fig['data'] += [bind(trace1, 1, 4)]
py.iplot(fig, filename='get_subplots-shared-bottom')
```

```python
fig = tls.make_subplots(rows=1, cols=4, 
                       shared_yaxes=True,  # sharing y-axes also works!
                       print_grid=True)
```

```python
fig['data'] += [bind(trace1, 1, 1)]
fig['data'] += [bind(trace1, 2, 1)]
fig['data'] += [bind(trace1, 3, 1)]
fig['data'] += [bind(trace1, 4, 1)]
py.iplot(fig, filename='get_subplots-shared2')
```

```python
fig = tls.make_subplots(rows=3, cols=3, 
                       shared_xaxes=True,  # more complicated layout
                       print_grid=True)
```

```python
fig['data'] += [bind(trace1, 1, 1)]
fig['data'] += [bind(trace1, 2, 2)]
fig['data'] += [bind(trace1, 3, 3)]
fig['data'] += [bind(trace1, 1, 4)]
fig['data'] += [bind(trace1, 2, 5)]
fig['data'] += [bind(trace1, 3, 6)]
fig['data'] += [bind(trace1, 1, 7)]
fig['data'] += [bind(trace1, 2, 8)]
fig['data'] += [bind(trace1, 3, 9)]
py.iplot(fig, filename='get_subplots-shared2a')
```

```python
fig = tls.make_subplots(rows=3, cols=3, 
                        shared_xaxes=True,
                        shared_yaxes=True,  # axis sharing in x & y 
                        print_grid=True)
```

```python
fig['data'] += [bind(trace1, 1, 1)]
fig['data'] += [bind(trace1, 2, 1)]
fig['data'] += [bind(trace1, 3, 1)]
fig['data'] += [bind(trace1, 1, 2)]
fig['data'] += [bind(trace1, 2, 2)]
fig['data'] += [bind(trace1, 3, 2)]
fig['data'] += [bind(trace1, 1, 3)]
fig['data'] += [bind(trace1, 2, 3)]
fig['data'] += [bind(trace1, 3, 3)]
py.iplot(fig, filename='get_subplots-shared2b')
```

A layout like this https://plot.ly/python/subplots/#Subplots-with-Shared-Axes can be easily made using `make_subplots`

```python
fig = tls.make_subplots(rows=2, cols=2,
                       shared_xaxes=[(1,1), (2,1)],  # (1,1) == top-left cell
                       shared_yaxes=[(1,1), (1,2)],
                       print_grid=True)
```

```python
fig['data'] += [bind(trace1, 1, 1)]
fig['data'] += [bind(trace1, 2, 1)]
fig['data'] += [bind(trace1, 1, 2)]
fig['data'] += [bind(trace1, 3, 3)]
py.iplot(fig, filename='get_subplots-shared3')
```

### Custom layout using subplot specs

Inspired by matplotlib's [gridspec](http://matplotlib.org/users/gridspec.html).

Here are a few examples:

```python
fig = tls.make_subplots(rows=3, cols=3,
                        specs=[[{}, {}, {'rowspan': 2}],
                               [{'colspan': 2}, None, None],
                               [{'colspan': 3}, None, None]],
                        # specs w/ 'colspan' & 'rowspan'
                        print_grid=True)
```

```python
fig['data'] += [bind(trace1, 1, 1)]
fig['data'] += [bind(trace1, 2, 2)]
fig['data'] += [bind(trace1, 3, 3)]
fig['data'] += [bind(trace1, 4, 4)]
fig['data'] += [bind(trace1, 5, 5)]
py.iplot(fig, filename='get_subplots-specs')
```

```python
fig = tls.make_subplots(rows=3, cols=3,
                        specs=[[{'rowspan': 2, 'colspan': 2}, None, {}],
                               [None, None, {}],
                               [{}, {'colspan': 2}, None]],
                        print_grid=True)
```

```python
fig['data'] += [bind(trace1, 1, 1)]
fig['data'] += [bind(trace1, 2, 2)]
fig['data'] += [bind(trace1, 3, 3)]
fig['data'] += [bind(trace1, 4, 4)]
fig['data'] += [bind(trace1, 5, 5)]
py.iplot(fig, filename='get_subplots-specs2')
```

```python
fig = tls.make_subplots(rows=3, cols=3,
                        specs=[[{'rowspan': 3}, {}, {}],
                               [None, {}, {}],
                               [None, {'colspan': 2}, None]],
                        print_grid=True)
```

```python
fig['data'] += [bind(trace1, 1, 1)]
fig['data'] += [bind(trace1, 2, 2)]
fig['data'] += [bind(trace1, 3, 3)]
fig['data'] += [bind(trace1, 4, 4)]
fig['data'] += [bind(trace1, 5, 5)]
fig['data'] += [bind(trace1, 6, 6)]
py.iplot(fig, filename='get_subplots-specs4')
```

```python
fig = tls.make_subplots(rows=2, cols=2,
                        horizontal_spacing=0, vertical_spacing=0,
                        specs=[[{'l': 0.1}, {'b': 0.2}],
                               [{'t': 0.2}, {'r': 0.1}]],  # custom padding !
                        print_grid=True)
```

```python
fig['data'] += [bind(trace1, 1, 1)]
fig['data'] += [bind(trace1, 2, 2)]
fig['data'] += [bind(trace1, 3, 3)]
fig['data'] += [bind(trace1, 4, 4)]
py.iplot(fig, filename='get_subplots-specs5')
```

### Insets 

```python
fig = tls.make_subplots(rows=2, 
                        insets=[{'cell': (0,0), 'l': 0.7, 'b': 0.2},
                                {'cell': (1,0), 'l': 0.7, 'b': 0.2}],
                        print_grid=True)
```

```python
fig['data'] += [bind(trace1, 1, 1)]
fig['data'] += [bind(trace1, 2, 2)]
fig['data'] += [bind(trace1, 3, 3)]
fig['data'] += [bind(trace1, 4, 4)]
py.iplot(fig, filename='get_subplots-insets')
```

```python
fig['data'] += [bind(trace1, 1, 1)]
fig['data'] += [bind(trace1, 2, 2)]
fig['data'] += [bind(trace1, 3, 3)]
fig['data'] += [bind(trace1, 4, 4)]
```

```python
py.iplot(fig, filename='get_subplots-insets2')
```
