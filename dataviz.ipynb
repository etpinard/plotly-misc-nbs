### Original matplotlib figure

First, remake the original matplotlib figure. The following code cell was opied verbatim from the Randal S. Olsen's blog <a href="" target="_blank">post</a> with the exception of the last line where we grab the underlying figure object and link it to a variable named `dataviz1`.

```python
%pylab inline  
from pandas import read_csv  
  
# Read the data into a pandas DataFrame.  
gender_degree_data = read_csv("http://www.randalolson.com/wp-content/uploads/"
                              "percent-bachelors-degrees-women-usa.csv")  
  
# These are the "Tableau 20" colors as RGB.  
tableau20 = [(31, 119, 180), (174, 199, 232), (255, 127, 14), (255, 187, 120),  
             (44, 160, 44), (152, 223, 138), (214, 39, 40), (255, 152, 150),  
             (148, 103, 189), (197, 176, 213), (140, 86, 75), (196, 156, 148),  
             (227, 119, 194), (247, 182, 210), (127, 127, 127), (199, 199, 199),  
             (188, 189, 34), (219, 219, 141), (23, 190, 207), (158, 218, 229)]  
  
# Scale the RGB values to the [0, 1] range, which is the format matplotlib accepts.  
for i in range(len(tableau20)):  
    r, g, b = tableau20[i]  
    tableau20[i] = (r / 255., g / 255., b / 255.)  
  
# You typically want your plot to be ~1.33x wider than tall. This plot is a rare  
# exception because of the number of lines being plotted on it.  
# Common sizes: (10, 7.5) and (12, 9)  
figure(figsize=(12, 14))  
  
# Remove the plot frame lines. They are unnecessary chartjunk.  
ax = subplot(111)  
ax.spines["top"].set_visible(False)  
ax.spines["bottom"].set_visible(False)  
ax.spines["right"].set_visible(False)  
ax.spines["left"].set_visible(False)  
  
# Ensure that the axis ticks only show up on the bottom and left of the plot.  
# Ticks on the right and top of the plot are generally unnecessary chartjunk.  
ax.get_xaxis().tick_bottom()  
ax.get_yaxis().tick_left()  
  
# Limit the range of the plot to only where the data is.  
# Avoid unnecessary whitespace.  
ylim(0, 90)  
xlim(1968, 2014)  
  
# Make sure your axis ticks are large enough to be easily read.  
# You don't want your viewers squinting to read your plot.  
yticks(range(0, 91, 10), [str(x) + "%" for x in range(0, 91, 10)], fontsize=14)  
xticks(fontsize=14)  
  
# Provide tick lines across the plot to help your viewers trace along  
# the axis ticks. Make sure that the lines are light and small so they  
# don't obscure the primary data lines.  
for y in range(10, 91, 10):  
    plot(range(1968, 2012), [y] * len(range(1968, 2012)), "--", 
         lw=0.5, color="black", alpha=0.3)  
  
# Remove the tick marks; they are unnecessary with the tick lines we just plotted.  
plt.tick_params(axis="both", which="both", bottom="off", top="off",  
                labelbottom="on", left="off", right="off", labelleft="on")  
  
# Now that the plot is prepared, it's time to actually plot the data!  
# Note that I plotted the majors in order of the highest % in the final year.  
majors = ['Health Professions', 'Public Administration', 'Education', 'Psychology',  
          'Foreign Languages', 'English', 'Communications\nand Journalism',  
          'Art and Performance', 'Biology', 'Agriculture',  
          'Social Sciences and History', 'Business', 'Math and Statistics',  
          'Architecture', 'Physical Sciences', 'Computer Science',  
          'Engineering']  
  
for rank, column in enumerate(majors):  
    # Plot each line separately with its own color, using the Tableau 20  
    # color set in order.  
    plot(gender_degree_data.Year.values,  
            gender_degree_data[column.replace("\n", " ")].values,  
            lw=2.5, color=tableau20[rank])  
      
    # Add a text label to the right end of every line. Most of the code below  
    # is adding specific offsets y position because some labels overlapped.  
    y_pos = gender_degree_data[column.replace("\n", " ")].values[-1] - 0.5  
    if column == "Foreign Languages":  
        y_pos += 0.5  
    elif column == "English":  
        y_pos -= 0.5  
    elif column == "Communications\nand Journalism":  
        y_pos += 0.75  
    elif column == "Art and Performance":  
        y_pos -= 0.25  
    elif column == "Agriculture":  
        y_pos += 1.25  
    elif column == "Social Sciences and History":  
        y_pos += 0.25  
    elif column == "Business":  
        y_pos -= 0.75  
    elif column == "Math and Statistics":  
        y_pos += 0.75  
    elif column == "Architecture":  
        y_pos -= 0.75  
    elif column == "Computer Science":  
        y_pos += 0.75  
    elif column == "Engineering":  
        y_pos -= 0.25  
      
    # Again, make sure that all labels are large enough to be easily read  
    # by the viewer.  
    text(2011.5, y_pos, column, fontsize=14, color=tableau20[rank])  
      
# matplotlib's title() call centers the title on the plot, but not the graph,  
# so I used the text() call to customize where the title goes.  
  
# Make the title big enough so it spans the entire plot, but don't make it  
# so big that it requires two lines to show.  
  
# Note that if the title is descriptive enough, it is unnecessary to include  
# axis labels; they are self-evident, in this plot's case.  
text(1995, 93, "Percentage of Bachelor's degrees conferred to women in the U.S.A."  
       ", by major (1970-2012)", fontsize=17, ha="center")  
  
# Always include your data source(s) and copyright notice! And for your  
# data sources, tell your viewers exactly where the data came from,  
# preferably with a direct link to the data. Just telling your viewers  
# that you used data from the "U.S. Census Bureau" is completely useless:  
# the U.S. Census Bureau provides all kinds of data, so how are your  
# viewers supposed to know which data set you used?  
text(1966, -8, "Data source: nces.ed.gov/programs/digest/2013menu_tables.asp"  
       "\nAuthor: Randy Olson (randalolson.com / @randal_olson)"  
       "\nNote: Some majors are missing because the historical data "  
       "is not available for them", fontsize=10)  
  
# Finally, save the figure as a PNG.  
# You can also save it as a PDF, JPEG, etc.  
# Just change the file extension in this call.  
# bbox_inches="tight" removes all the extra whitespace on the edges of your plot.  
#savefig("percent-bachelors-degrees-women-usa.png", bbox_inches="tight");  

# (!) Grab figure object and link it to variable (must be in same cell as figure)
dataviz1 = gcf()
```

Plotly allows you to convert a matplotlib figure object (`dataviz1` in our case here) into a Plotly figure with one line of code:

```python
py.iplot_mpl(dataviz1, filename='dataviz1')
```

Hmm. 

Unfortunately, the matplotlib to Plotly <a href="https://github.com/mpld3/matplotlylib" target="_blank">converter</a> isn't perfect yet.

But luckily, Plotly's Python API makes it easy to add back features missed during the conversion.

So, our tasks are:

* [1.1 Resize the figure](#1.1-Resize-the-figure)

to the same aspect ratio as the original.

* [1.2 Remove frame and ticks, add Plotly grid lines](#1.2-Remove-frame-and-ticks,-add-Plotly-grid-lines)

Dashed grid lines are currently not supported in Plotly; so the grid lines were erroneously converted to traces.

* [1.3 Adjust annotations, add Plotly title](#1.3-Adjust-annotations,-add-Plotly-title)

Annotations are poorly placed in the Plotly figure. Newlines in Plotly are set with the `<br>` HTML tag instead of the `\n ` escape sequence. Moreover, we will add a title to make Plotly figure's URLs more descriptive.

* [1.4 Add hover notes to data](#1.4-Add-hover-text-to-data)

To make full use of Plotly's interactibility.

### Plotly version

So, first let's convert the matplotlib figure object to a Plotly figure object:

```python
dataviz1_plotly = tls.mpl_to_plotly(dataviz1)

print dataviz1_plotly.to_string()   # show plotly figure object in notebook
```

##### 1.1 Resize the figure 

```python
width = 750           # width of figure in pixels (to fit in notebook)
ratio = 12./14        # same aspect ratio as original figure
height = width/ratio  # height of figure in pixels

# Update the layout object
dataviz1_plotly['layout'].update(
    autosize=False,     # must be set to False for custom sizing
    width=width,
    height=height
)
```

##### 1.2 Remove frame and ticks, add Plotly grid lines

Update each axis:

```python
# Update x-axis
dataviz1_plotly['layout']['xaxis1'].update(
    showline=False,  # remove axis line
    ticks=''         # remove all ticks
)

# Update y-axis
dataviz1_plotly['layout']['yaxis1'].update(
    showline=False,
    ticks='',
    showgrid=True    # show horizontal grid lines
)
```

Update data object so that the erroneously converted grid line do not show on the figure:

```python
# List of traces in data object
traces = dataviz1_plotly['data']

# 'visible': False if dashed, True otherwise
flags_dash = [trace['line']['dash']!='dash' for trace in traces]
update_dash = [{'visible': flag} for flag in flags_dash]

# Update each trace in data object
dataviz1_plotly['data'].update(update_dash)
```

Plotly only shows grid line **outside** the a given axis' range. Therefore, adjust the y-axis' range so that the $y=0$ and $y=90$ grid lines appear as in the original figure:

```python
dataviz1_plotly['layout']['yaxis1'].update(range=[-0.1,90.1])
```

##### 1.3 Adjust annotations, add Plotly title

```python
# Show annotation objects in notebook
print dataviz1_plotly['layout']['annotations'].to_string() 
```

Notice that annotations containing the names of each major have the same x coordinate as in the original (i.e. $x=2011.5$). The converter did its job there.

That said, looking at Plotly figure the annotations' text is **centered** at that x coordinate whereas, in the original, the annotations' text start **to the right of** the given x coordinate. 

So consider,

```python
# Update every annotations
dataviz1_plotly['layout']['annotations'].update(dict(
    xanchor='left',    #  set anchor to the left of the text
    yanchor='bottom'   #  set anchor below the text
))
```

Next, replace all the `\n ` sequence escape to `<br>` HTML tags:

```python
# List of all annotation objects
annos = dataviz1_plotly['layout']['annotations']

# List of all annotation texts, show it in notebook
annos_text = [anno['text'] for anno in annos]
annos_text
```

```python
# Replace \n to <br> 
annos_text_updated = [text.replace('\n','<br>') for text in annos_text]
```

```python
# List of majors (for task 1.4)
majors = annos_text_updated[:-2]

# Make the title span 2 lines 
title = annos_text_updated[-2].replace('in','<br>in')

# Text for the source annotations
source = annos_text_updated[-1]
```

```python
# Link updated annotation text to 'text' key for all annotation
update_anno_text = [{'text': text } for text in majors+[' ']+[source]]

# Update annotations
dataviz1_plotly['layout']['annotations'].update(update_anno_text)
```

```python
# Add title
dataviz1_plotly['layout'].update(
    title=title,
    titlefont=dict(size=20)  # increase font size
)
```

```python
# Adjust margins 
dataviz1_plotly['layout']['margin'].update(
    l=50,   # left margin in pixels
    r=160,  # right " " "
    b=100,  # bottom " " "
    t=100   # top " " "
)
```

In Plotly annotations outside the span of the axes are not shown. So, we need to set the *source* annotation to grid coordinates. 

```python
# Show 'source' annotation object in notebook
print dataviz1_plotly['layout']['annotations'][-1].to_string()
```

```python
# Update 'source' annotation
dataviz1_plotly['layout']['annotations'][-1].update(
    xref='paper',   # use paper (not axis) coordinates for x position
    yref='paper',   # " " " " " " y position
    x=-0.01,      # x position in normalized coordinates
    y=-0.11,      # y position " " "
    align='left'  # left-aligned text
)
```

##### 1.4 Add hover notes to data

Recall that `dataviz1_plotly` contains erronerously converted grid lines. So, get the indices corresponding the actual traces:

```python
# List of indices corresponding to traces (from task 1.2)
trace_indices = [i_flag for i_flag, flag in enumerate(flags_dash) if flag]
```

Then,

```python
# Link major name to 'name' key for each corresp. trace
update_name = [{'name': major} for major in majors]

# Update each trace 
for i_text, i_trace in enumerate(trace_indices):
    dataviz1_plotly['data'][i_trace].update(update_name[i_text])
```

Finally, make every y coordinate show when hovering over a given x coordinate:

```python
dataviz1_plotly['layout'].update(hovermode='x')
```

### Send updated Plotly figure object to Plotly!

```python
py.iplot(dataviz1_plotly, filename='dataviz1_updated', 
         width=width, height=int(height))
```
