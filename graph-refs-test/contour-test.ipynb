### Heatmap

```python
help(Heatmap)
```

```python
my_data = ([Heatmap(z=[[1, 20, 30], 
                       [20, 1, 60], 
                       [30, 60, 1]],
                    zauto=False,
                    zmax=40,
                    showscale=False,
            )])

#my_layout = Layout(yaxis= YAxis(autorange='reversed'))
my_layout = Layout(title='testing Heatmap keys')

my_fig = Figure(data=my_data, layout=my_layout)
py.iplot(my_fig, filename='test-heatmap-keys')
```

### Contour

```python
help(Contour)
```

```python
my_data = ([Contour(z=[[1, 20, 30], 
                       [20, 1, 60], 
                       [30, 60, 1]],
                    #x0=-100,
                    dx=0.1,
                    dy=2
                    
            )])

my_layout = Layout(title='testing Contour keys')

my_fig = Figure(data=my_data, layout=my_layout)
py.iplot(my_fig, filename='test-contour-keys')
```

### Histogram 2d

```python
help(Histogram2d)
```

```python
np.random.exponential(size=10)
```

```python
my_data = ([Histogram2d(x = np.random.randn(100),
                        y = np.random.exponential(size=100),
                        #zmax=30
                        #marker= Marker(color='white',),
                        line= Line(color='white',
                                   width=10)
                    
            )])

my_layout = Layout(title='testing Histogram2d keys')

my_fig = Figure(data=my_data, layout=my_layout)
py.iplot(my_fig, filename='test-histogram2d-keys')
```

### Histogram2dContour

```python
help(Histogram2dContour)
```

```python
my_data = ([Histogram2dContour(x = np.random.randn(100),
                               y = np.random.exponential(size=100),
                               zauto=False,
                               #zmax=30
                    
                              )])

my_layout = Layout(title='testing Histogram2dContour keys')

my_fig = Figure(data=my_data, layout=my_layout)
py.iplot(my_fig, filename='test-histogram2dcontour-keys', validate=False)
```
