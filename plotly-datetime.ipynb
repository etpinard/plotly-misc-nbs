```python
import plotly.plotly as py
import plotly.tools as tls
from plotly.graph_objs import *
```

```python
timestamp = ['2014-11-10 07:59:15', '2014-11-10 08:59:15', 
             '2014-11-10 09:59:15', '2014-11-10 10:59:15',
             '2014-11-10 11:59:15', '2014-11-10 12:59:15',
             '2014-11-10 13:59:16', '2014-11-10 14:59:15',
             '2014-11-10 15:59:16', '2014-11-10 16:59:16',
             '2014-11-10 17:59:15', '2014-11-10 18:59:15',
             '2014-11-10 19:59:15', '2014-11-10 20:59:16']
temp = [16.4, 16.3, 16.1, 16.1, 16.2, 16.2, 16.1, 16.1,
        16.0, 15.9, 15.8, 15.7, 15.6, 15.5]
```

```python
data = Data([
    Scatter(
        x=timestamp,
        y=temp
    )
])
py.iplot(data, filename='python-datetime')
```
